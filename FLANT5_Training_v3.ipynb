Convert the dataset to jsonl format


```python
import json

def convert_dataset(data):
    new_data = []
    for item in data:
        input_text = item["input"]
        output_text = item["output"]

        new_data.append({
            "input": input_text,
            "output": output_text
        })
    return new_data

with open("newdatav3.json", "r") as f:
    original_data = json.load(f)

converted_data = convert_dataset(original_data)

with open("newdatav3.jsonl", "w") as f:
    for item in converted_data:
        f.write(json.dumps(item) + "\n")

print("✅ Conversion successful! Saved as 'newdatav3.jsonl'")
```

```python
dataset = load_dataset("json", data_files={"train": "/content/newdatav3.jsonl"})
```

Tokenize the dataset


```python
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
```

Split data into training and validation

```python
from datasets import DatasetDict

dataset = dataset["train"].train_test_split(test_size=0.1, seed=42)

dataset = DatasetDict({
    "train": dataset["train"],
    "validation": dataset["test"]
})
```

Preprocessing the dataset which returns a dictionary with input Id and labels

```python
from transformers import AutoTokenizer

max_input_length = 128
max_target_length = 384

def preprocess_function(examples):
    inputs = [ex for ex in examples["input"]]
    targets = [ex for ex in examples["output"]]
    model_inputs = tokenizer(
        inputs,
        max_length=max_input_length,
        truncation=True,
        padding="max_length",
    )
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            targets, max_length=max_target_length, truncation=True, padding="max_length"
        )
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

Train the model and compile

```python
from transformers import TFAutoModelForSeq2SeqLM, DataCollatorForSeq2Seq, create_optimizer
import tensorflow as tf

# Load pre-trained model
model = TFAutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")

# Training parameters
batch_size = 8
num_epochs = 5
learning_rate = 3e-5

# Create data collator
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf")
# 'tokenizer' should be defined earlier in your notebook

# Create TensorFlow datasets
tf_train_dataset = tokenized_datasets["train"].to_tf_dataset(
    columns=["input_ids", "attention_mask", "labels"],
    shuffle=True,
    batch_size=batch_size,
    collate_fn=data_collator,
)
tf_validation_dataset = tokenized_datasets["validation"].to_tf_dataset(
    columns=["input_ids", "attention_mask", "labels"],
    shuffle=False,
    batch_size=batch_size,
    collate_fn=data_collator,
)

# Calculate training steps
num_train_steps = len(tf_train_dataset) * num_epochs

# Create optimizer
optimizer, schedule = create_optimizer(
    init_lr=learning_rate,
    num_warmup_steps=0,
    num_train_steps=num_train_steps,
)

# Compile model
model.compile(optimizer=optimizer, metrics=["accuracy"])

# Train model
model.fit(tf_train_dataset, epochs=num_epochs, validation_data=tf_validation_dataset)

# Save fine-tuned model
model.save_pretrained("fine_tuned_flan_t5")
```
